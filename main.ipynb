# Import library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Grafik
Grafik adalah penyajian data dalam bentuk visual (gambar, garis, batang, lingkaran) untuk menyederhanakan informasi kuantitatif agar lebih mudah dipahami, menunjukkan hubungan antar variabel, tren, atau perbandingan secara ringkas dan jelas daripada membaca angka mentah. Grafik mengubah data numerik menjadi gambar yang intuitif, membantu melihat pola dan kesimpulan dengan cepat.

```python
x = np.array([2020,2021,2022,2023,2024,2025,2026])
y1 = np.array([13,4,12,45,12,56,32])
y2 = np.array([17,34,25,39,45,67,22])
y3 = np.array([14,40,71,88,32,24,24])

line_style = dict(marker="o",
               markersize=7,                # bisa disingkat dengan 'ms'
               markerfacecolor="#000000", # bisa disingkat dengan 'mfc'
               markeredgecolor="#000000", # bisa disingkat dengan 'mec'
               linestyle="solid",
               linewidth=3)

plt.title("Class size", fontsize=25,
                        family="sans-serif",
                        fontweight="bold",
                        color="#2d4cfc",)    # mengatur judul grafik

plt.xlabel("Year", fontsize=20,
                   family="sans-serif",
                   fontweight="bold",
                   color="#2dbefc")         # mengatur nama sumbu x

plt.ylabel("Students", fontsize=20,
                       family="sans-serif",
                       fontweight="bold",
                       color="#2dbefc")     # mengatur nama sumbu y

plt.tick_params(axis="x",
                colors="#ff0000")           # kustom angka

plt.plot(x, y1, color="#1c5bfc", label="Class A", **line_style)
plt.plot(x, y2, color="#1cfc45", label="Class B", **line_style)
plt.plot(x, y3, color="#fc1c1c", label="Class C", **line_style)    # menampilkan 3 data dalam satu grafik

plt.legend(title="Daftar Kelas", 
           title_fontsize=12, 
           loc="upper left",    # Posisi legenda (bisa: upper right, lower center, dll)
           fontsize=10, 
           shadow=True,         # Kasih efek bayangan biar keren
           facecolor="white")

plt.grid(axis="y",
         linewidth="3",
         color="lightgray",
         linestyle="dotted")                        # kustomisasi grid

plt.xticks(x)
plt.show()
```

# Diagram batang / bar chart
Diagram batang adalah teknik visualisasi data yang menggunakan batang-batang persegi panjang (vertikal atau horizontal) untuk menampilkan dan membandingkan nilai atau frekuensi dari kategori data diskret atau kualitatif, di mana panjang atau tinggi batang tersebut proporsional dengan nilainya, memudahkan pemahaman pola dan perbandingan data secara cepat. 

```python
categories = np.array(["Grains", "Fruits", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4,5,2,7,4,3])

plt.bar(categories, values, color="skyblue")
#plt.barh(categories, values, color="skyblue")

plt.title("Daily consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

# Diagram lingkaran / pie chart
Diagram lingkaran (atau pie chart) adalah cara penyajian data statistik dengan menggunakan gambar berbentuk lingkaran yang dibagi menjadi beberapa bagian atau juring. 


```python
categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",
                colors=colors,
                explode=[0, 0, 0, 0.2],
                shadow=True,
                startangle=90)

plt.title("Politeknik Elektronika Negeri Surabaya")

plt.show()
```

# Diagram sebar / scatter diagram
Diagram sebar (juga dikenal sebagai scatter plot, scatter diagram, atau diagram pencar) adalah grafik yang digunakan untuk menunjukkan hubungan atau korelasi antara dua variabel numeri

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]) # jam belajar
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]) # nilai

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8]) # jam belajar
y2 = np.array([50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97]) # nilai

plt.scatter(x1, y1, color="skyblue",
                  alpha=0.5,
                  s=200,
                  label="Class A")

plt.scatter(x2, y2, color="red",
                  alpha=0.5,
                  s=200,
                  label="Class B")

plt.title("Test scores")
plt.xlabel("Hours studied")
plt.ylabel("Grade")

plt.legend()
plt.show()
```

# Histogram
Histogram adalah diagram batang yang digunakan untuk menggambarkan distribusi frekuensi dari suatu kumpulan data numerik berkelanjutan (continuous data)


```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")

plt.title("Exam scores")
plt.xlabel("Score")
plt.ylabel("# of student")

plt.show()
```

# Subplot
Figure = The entire canvas  
Ax = A single plot (subplot)

```python
x = np.array([1,2,3,4,5])

figure, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, x*2, color="red")
axes[0, 0].set_title("X*2")

axes[0, 1].bar(x, x**2, color="blue")
axes[0, 1].set_title("X**2")

axes[1, 0].plot(x, x**3, color="green")
axes[1, 0].set_title("X**3")

axes[1, 1].plot(x, x**4, color="black")
axes[1, 1].set_title("X**4")

plt.tight_layout()
plt.show()
```
